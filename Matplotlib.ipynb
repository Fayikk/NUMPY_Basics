##MATPLOTLİB
Matplotlib, Python programlama dili ve sayısal matematik uzantısı NumPy için bir çizim kitaplığıdır.

```python
import matplotlib.pyplot as plt
import numpy as np
"""ÖRNEK"""
x=[1,2,3,4] # x ekseni için girlen değerler
y=[1,4,9,16] #y ekseni için girlen değerler
plt.plot(x,y,"--*") #x ve y ekseni için değerlerin okunabilmesi(grafiğin çizimi) için gerekli karakterler.
plt.axis([0,6,0,20]) # x ve y ekseni için her iki doğrultunun başlangıç ve bitiş noktasını burada ayarlamaktayız
plt.title("Grafik başlığı") #grafiğin başlığını buradan ayarlarız.
plt.xlabel("speed") #grafiğin etiketini buradan ayarlarız(x ekseninin tuttuğu değerin olgu)
plt.ylabel("time") # y ekseninin belirttiği olgu
plt.show() #grafiği çalıştırmak için kod parçası.


```

```python
"""ÖRNEK 2"""

# z=np.linspace(0,2,100)
# print(z)

x=np.linspace(0,2,100)
plt.plot(x,x,label="linear",color="orange")
plt.plot(x,x**2,label="quadratic",color="red")
plt.plot(x,x**3,label="cubic",color="black")

plt.xlabel("empty")
plt.ylabel("empty")
plt.title("basics plot")
plt.show()
```

```python
"""ÖRNEK3 PİE(PASTA)GRAPH"""
goal_types="Penaltı","Kaleye atılan şut","Serbest Vuruş"
goals=[12,35,7]
colors=['y','r','b']
plt.pie(goals,labels=goal_types,colors=colors,shadow=True,explode=(0.05,0.05,0.05))
plt.show()
```

#GRAFİĞİMİZ OLUŞTURALIM

```python
import matplotlib.pyplot as plt
import numpy as np

x=[1,2,3,4,5,6,7,8,9]
y=[9,8,7,6,5,4,3,2,1]
plt.plot(x,y,"o--r",linewidth=3) #grafiğimizi oluşturmaya plot() metodu ile başlıyoruz
plt.axis([0,6,0,20])
plt.title("DENEME GRAFİĞİ")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")

plt.show()
```

```python
x=np.linspace(0,2,100)
plt.plot(x,x,label="Linear",color="purple")
plt.plot(x,x**2,label="Quadratic")
plt.plot(x,x**3,label="Cubic")
plt.legend() #grafiğin sol üzerindeki isimlerin yazılmasını sağlıyordur
plt.title("THİS İS GRAPHİC TİTLE")
plt.xlabel("X axis")
plt.ylabel("Y axis")


plt.show()
```

```python
x=np.linspace(0,2,100)

fig,axs=plt.subplots(3)
axs[0].plot(x,x,color="red")
axs[1].plot(x,x**2,color="black")
axs[2].plot(x,x**3,color="blue")

plt.tight_layout()#karışan grafikleri düzgün bir layout düzene sokma işine yarar.
plt.show()
```

```python
#Burada cube parabolü bulunmaktadır.
x=np.linspace(-10,9,20)
y=x**3
# fig,axs=plt.subplots() #bu satırda figür oluşturuyoruz.
figure=plt.figure()
axes_cube=figure.add_axes([0.1,0.1,0.9,0.9])
axes_cube.plot(x,y,'b')
axes_cube.set_xlabel("X AXİS")
axes_cube.set_ylabel("Y AXİS")
axes_cube.set_title("Cube")
plt.show()
```
